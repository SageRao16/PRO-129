```python
import pandas as pd
```

```python
bright_stars_df = pd.read_csv('/content/PRO128/bright_stars.csv')
dwarf_stars_df = pd.read_csv('/content/PRO128/dwarf_stars.csv')
```

```python
bright_stars_df.head()
```

```python
dwarf_stars_df.head()
```

```python
dwarf_stars_df = dwarf_stars_df.dropna()
```

```python
dwarf_stars_df.head()
```

```python
dwarf_stars_df.Mass = dwarf_stars_df.Mass.replace('[^a-zA-Z0-9]', '').astype('float')
```

```python
dwarf_stars_df['Mass'] = dwarf_stars_df['Mass']*0.000954588
```

```python
dwarf_stars_df['Radius'] = dwarf_stars_df['Radius']*0.102763
```

```python
dwarf_stars_df.head()
```

```python
merge_planets = pd.merge(dwarf_stars_df,bright_stars_df,on = 'id')
merge_planets.shape
merge_planets.to_csv('129_final.csv')
```
